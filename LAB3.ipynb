```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

Load Dataset


```python
import kagglehub
import os

path = kagglehub.dataset_download("akashdeepkuila/bakery")



print(path)
```

```python
csv_path = os.path.join(path, "Bakery.csv")
df = pd.read_csv(csv_path, encoding="latin-1")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Items'], bins=20)
plt.title("Distribution of items")
plt.show()
```

```python
items_sales_counts = df['Items'].value_counts()

plt.figure(figsize=(10,5))
items_sales_counts.head(20).plot(kind='bar')
plt.title("Top 20 Items by Number of Sales")
plt.ylabel("Number of Sales")
plt.xlabel("Item")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Top 10 items by sales count:")
print(items_sales_counts.head(10))
```
